```python
from pydataset import data
swiss=data('swiss')
swiss
```

```python
swiss['is_catholic']=swiss.Catholic>50
```

```python
swiss
```

```python
sns.catplot(x='is_catholic',y='Fertility', data=swiss)
```

```python
sns.relplot(y='Infant.Mortality',x='Fertility', data=swiss)
sns.relplot(y='Education',x='Fertility', data=swiss)
sns.relplot(y='Examination',x='Fertility', data=swiss)
sns.relplot(y='Agriculture',x='Fertility', data=swiss)
```

```python
labels = ['Fertility', 'Agruculture', 'Examination', 'Infant.Mortality', 'Education','Catholic','is_catholic']
sns.heatmap(swiss.corr(),
           annot=True,
           cmap='coolwarm',
           vmax=1,
           vmin=-1,
           center=0,
           linewidth=1,
           linecolor='k',
           fmt='.2f',
           xticklabels=labels,
           yticklabels=labels)
```

```python
import env
user=env.user
host=env.host
password=env.password
url=f'mysql+pymysql://{user}:{password}@{host}/chipotle'
def get_db_url(n):
    n=url
    return url
```

```python
orders=pd.read_sql('SELECT * FROM orders',url)
```

```python
orders.info()
```

```python
orders['total_prices']=(orders.item_price.str.replace('[$,]','').astype(float))*orders.quantity
orders
```

```python
top=orders.groupby('item_name').total_prices.sum().nlargest(4)
top_4=pd.DataFrame(data=top, columns=['item_name', 'total_prices'])
top_4.item_name=top_4.index
top_4
```

```python
sns.barplot(data=top_4,x='item_name',y='total_prices')
```
